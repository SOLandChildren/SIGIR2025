```python
import pandas as pd

ql_it = pd.read_csv("../Data/Queries/QL_IT.csv")
ql_it_em = pd.read_csv("../Data/Queries/QL_IT_Em.csv")
```

```python
sent_dict = {
    "Bad": "Negative",
    "Good": "Positive"
}

def get_task_sentiment(query):
    sent = ql_it_em.loc[ql_it_em["user_query"]==query]["target"].to_list()[0]
    return sent_dict[sent]
```

```python
ql_it_queries = ql_it["user_query"].unique().tolist()
ql_it_em_queries = ql_it_em["user_query"].unique().tolist()

final_queries = []
for gen_query, i in zip(ql_it_queries, range(1, len(ql_it_queries)+1)):
    qid = "QGEN" + str(i)
    final_queries.append([qid, gen_query, "General", None])

for em_query, i in zip(ql_it_em_queries, range(1, len(ql_it_em_queries)+1)):
    qid = "QEM"+str(i)
    final_queries.append([qid, em_query, "Emotionally-Charged", get_task_sentiment(em_query)])
```

```python
final_queries_df = pd.DataFrame(final_queries, columns=["QID", "Query", "Prompt Type", "Task Sentiment"])
final_queries_df.to_csv("../Data/Queries/Queries_final.csv", index=False)
final_queries_df
```
