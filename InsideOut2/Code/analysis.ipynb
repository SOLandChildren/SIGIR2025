## Importing files

```python
import pandas as pd

query_file_name = ""
query_file_path = "../Data/"+query_file_name
queries = pd.read_csv(query_file_path)

queries_lexicon = pd.read_csv("../Results/LexIT_queryWise_queryEP.csv")
queries_lexicon["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in queries_lexicon["QID"]]
queries_semantic = pd.read_csv("../Results/FEELIT_queryWise_queryEP.csv")
queries_semantic["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in queries_semantic["QID"]]

gemma_lexicon = pd.read_csv("../Results/LexIT_queryWise_gemmaEP.csv")
gemma_lexicon["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in gemma_lexicon["QID"]]
gemma_semantic = pd.read_csv("../Results/FEELIT_queryWise_gemmaEP.csv")
gemma_semantic["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in gemma_semantic["QID"]]

gpt_lexicon = pd.read_csv("../Results/LexIT_queryWise_gptEP.csv")
gpt_lexicon["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in gpt_lexicon["QID"]]
gpt_semantic = pd.read_csv("../Results/FEELIT_queryWise_gptEP.csv")
gpt_semantic["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in gpt_semantic["QID"]]

IO_lexicon = pd.read_csv("../Results/LexIT_queryWise_InsideOutEP.csv")
IO_semantic = pd.read_csv("../Results/FEELIT_queryWise_InsidOutEP.csv")

bing_lexicon = pd.read_csv("../Results/LexIT_queryWise_bingEP.csv")
bing_lexicon = bing_lexicon.loc[bing_lexicon["rank"]<6]
bing_lexicon["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in bing_lexicon["QID"]]

bing_semantic = pd.read_csv("../Results/FEELIT_queryWise_bingEP.csv")
bing_semantic = bing_semantic.loc[bing_semantic["rank"]<6]
bing_semantic["Task Sentiment"] = [queries.loc[queries["QID"]==qid]["Task Sentiment"].values[0] for qid in bing_semantic["QID"]]

QL_IT = pd.read_csv("../Data/InsideOutData/QL-IT.csv")
QL_IT_queries = pd.read_csv("../Data/InsideOutData/QL-IT_query.csv")
QL_IT_Em = pd.read_csv("../Data/InsideOutData/QL-IT-Em.csv")

QL_IT.columns = ['session_id', 'user_query', 'url', 'source', 'title', 'snippet',
       'position', 'TextSample', 'TextSample-Tokenized',
       'tagged_TextSample-Tokenized', 'anger', 'anticipation', 'disgust',
       'fear', 'joy', 'sadness', 'surprise', 'trust', 'positive',
       'negative', 'neutral']

QL_IT_queries.columns = ['session_id', 'user_query', 'url', 'source', 'title', 'snippet',
       'position', 'user_query-Tokenized', 'tagged_user_query-Tokenized',
       'anger', 'anticipation', 'disgust',
       'fear', 'joy', 'sadness', 'surprise', 'trust', 'positive',
       'negative', 'neutral']

QL_IT_Em.columns = ['session_id', 'Task Sentiment', 'user_query', 'url', 'source', 'title',
       'snippet', 'position', 'TextSample', 'TextSample-Tokenized',
       'tagged_TextSample-Tokenized', 'anger', 'anticipation', 'disgust',
       'fear', 'joy', 'sadness', 'surprise', 'trust', 'positive',
       'negative', 'neutral', 'user_query-Tokenized', 'tagged_user_query-Tokenized',
       'QE_anger', 'QE_anticipation', 'QE_disgust', 'QE_fear', 'QE_joy',
       'QE_sadness', 'QE_surprise', 'QE_trust', 'QS_pos', 'QS_neg', 'QS_obj']

# bringing QL_IT and QL_IT_Em emotion profile values to same 0-100 scale
for emo in [ 'anger', 'anticipation', 'disgust',
       'fear', 'joy', 'sadness', 'surprise', 'trust', 'positive',
       'negative', 'neutral']:
    QL_IT[emo] = [row[emo]*100 for _, row in QL_IT.iterrows()]
    QL_IT_Em[emo] = [row[emo]*100 for _, row in QL_IT_Em.iterrows()]
    QL_IT_queries[emo] = [row[emo]*100 for _, row in QL_IT_queries.iterrows()]
    
```

## Average emotional profiles

### Queries

```python
queries_gen = queries_lexicon.loc[queries_lexicon["QID"].str.startswith("qGEN")].mean(numeric_only=True).round(2).to_frame().T
queries_gen["Prompt Type"] = ["General"] * len(queries_gen)
queries_gen["Task Sentiment"] = ["Overall"] * len(queries_gen)
queries_em = queries_lexicon.loc[queries_lexicon["QID"].str.startswith("qEM")].mean(numeric_only=True).round(2).to_frame().T
queries_em["Prompt Type"] = ["EM"] * len(queries_em)
queries_em["Task Sentiment"] = ["Overall"] * len(queries_em)
queries_task = queries_lexicon.loc[queries_lexicon["QID"].str.startswith("qEM")].groupby("Task Sentiment").mean(numeric_only=True).round(2).reset_index()
queries_task["Prompt Type"] = ["EM"] * len(queries_task)
queries_lexicon_avg = pd.concat([queries_gen, queries_em, queries_task], ignore_index=True)
queries_lexicon_avg["Text Sample"] = ["Query"] * len(queries_lexicon_avg)
queries_lexicon_avg["Analysis"] = ["Lexical"] * len(queries_lexicon_avg)
queries_lexicon_avg
```

```python
queries_gen = queries_semantic.loc[queries_semantic["QID"].str.startswith("qGEN")].mean(numeric_only=True).round(2).to_frame().T
queries_gen["Prompt Type"] = ["General"] * len(queries_gen)
queries_gen["Task Sentiment"] = ["Overall"] * len(queries_gen)
queries_em = queries_semantic.loc[queries_semantic["QID"].str.startswith("qEM")].mean(numeric_only=True).round(2).to_frame().T
queries_em["Prompt Type"] = ["EM"] * len(queries_em)
queries_em["Task Sentiment"] = ["Overall"] * len(queries_em)
queries_task = queries_semantic.loc[queries_semantic["QID"].str.startswith("qEM")].groupby("Task Sentiment").mean(numeric_only=True).round(2).reset_index()
queries_task["Prompt Type"] = ["EM"] * len(queries_task)
queries_semantic_avg = pd.concat([queries_gen, queries_em, queries_task], ignore_index=True)
queries_semantic_avg["Text Sample"] = ["Query"] * len(queries_lexicon_avg)
queries_semantic_avg["Analysis"] = ["Semantic"] * len(queries_lexicon_avg)
queries_semantic_avg
```

```python
queries_avg = pd.concat([queries_lexicon_avg, queries_semantic_avg], ignore_index=True)
queries_avg = queries_avg.loc[:, ["Analysis", "Prompt Type", "Task Sentiment", "Text Sample"]+queries_lexicon.drop(["QID", "Task Sentiment"], axis=1).columns.tolist()]
queries_avg.to_csv("../Results/queries_avg_EP.csv", index=False)
queries_avg
```

## Stats test

```python
from scipy.stats import ttest_ind, f_oneway

emotions = ["positive", "negative", "neutral", "anger", "anticipation", "disgust", "fear", "joy", "sadness", "surprise", "trust"]

def stat_test(comp_name, data, num_comparisons):
    result = {
        "comp_name": comp_name
    }

    if num_comparisons == 2:
        for emo in emotions:
            if emo in data[0].columns:
                test_stat, p_val= ttest_ind(data[0][emo], data[1][emo],nan_policy='omit')
                result[emo] = float(p_val)
            else:
                result[emo]="N/A"
    elif num_comparisons == 3:
        for emo in emotions:
            if emo in data[0].columns:
                test_stat, p_val = f_oneway(data[0][emo], data[1][emo], data[2][emo], nan_policy='omit')
                result[emo] = float(p_val)
            else:
                result[emo]="N/A"
    elif num_comparisons == 5:
        for emo in emotions:
            if emo in data[0].columns:
                test_stat, p_val = f_oneway(data[0][emo], data[1][emo], data[2][emo], data[3][emo], data[4][emo], nan_policy="omit")
                result[emo] = float(p_val)
            else:
                result[emo]="N/A"

    return result
```

```python
def get_exp_stats(df, exp_num, df_name, is_se):
    if is_se == True:
        df_gen = df.loc[df["QID"].str.startswith("qGEN")]
        df_em = df.loc[df["QID"].str.startswith("qEM")]
        df_em_good = df.loc[df["Task Sentiment"]=="Good"]
        df_em_bad = df.loc[df["Task Sentiment"]=="Bad"]
        gen_em = {'exp':exp_num}|stat_test(comp_name=df_name+"_gen_em", data=[df_gen, df_em], num_comparisons=2)
        em_good_bad = {'exp':exp_num}|stat_test(comp_name=df_name+"_em_good_bad", data=[df_em_good, df_em_bad], num_comparisons=2)
        gen_rank = {'exp':exp_num}|stat_test(comp_name=df_name+"_gen_rank", data=[df_gen.loc[df_gen["rank"]==i] for i in range(1,6)], num_comparisons=5)
        em_rank = {'exp':exp_num}|stat_test(comp_name=df_name+"_em_rank", data=[df_em.loc[df_em["rank"]==i] for i in range(1,6)], num_comparisons=5)
        return gen_em, em_good_bad, gen_rank, em_rank
    else:
        gemma_df_gen = df.loc[(df["IAS"]=="Gemma") & (df["QID"].str.startswith("qGEN"))]
        gemma_df_em = df.loc[(df["IAS"]=="Gemma") & (df["QID"].str.startswith("qEM"))]
        gemma_df_em_good = df.loc[(df["IAS"]=="Gemma") & (df['Task Sentiment']=="Good")]
        gemma_df_em_bad = df.loc[(df["IAS"]=="Gemma") & (df['Task Sentiment']=="Bad")]
        
        gpt_df_gen = df.loc[(df["IAS"]=="GPT") & (df["QID"].str.startswith("qGEN"))]
        gpt_df_em = df.loc[(df["IAS"]=="GPT") & (df["QID"].str.startswith("qEM"))]
        gpt_df_em_good = df.loc[(df["IAS"]=="GPT") & (df['Task Sentiment']=="Good")]
        gpt_df_em_bad = df.loc[(df["IAS"]=="GPT") & (df['Task Sentiment']=="Bad")]
        
        bing_df_gen = df.loc[(df["IAS"]=="Bing") & (df["QID"].str.startswith("qGEN"))]
        bing_df_em = df.loc[(df["IAS"]=="Bing") & (df["QID"].str.startswith("qEM"))]
        bing_df_em_good = df.loc[(df["IAS"]=="Bing") & (df['Task Sentiment']=="Good")]
        bing_df_em_bad = df.loc[(df["IAS"]=="Bing") & (df['Task Sentiment']=="Bad")]

        IAS_gen = {'exp':exp_num}|stat_test(comp_name=df_name+"_gen", data=[gemma_df_gen, gpt_df_gen, bing_df_gen], num_comparisons=3)
        IAS_em = {'exp':exp_num}|stat_test(comp_name=df_name+"_em", data=[gemma_df_em, gpt_df_em, bing_df_em], num_comparisons=3)
        IAS_em_good = {'exp':exp_num}|stat_test(comp_name=df_name+"_em_good", data=[gemma_df_em_good, gpt_df_em_good, bing_df_em_good], num_comparisons=3)
        IAS_em_bad = {'exp':exp_num}|stat_test(comp_name=df_name+"_em_bad", data=[gemma_df_em_bad, gpt_df_em_bad, bing_df_em_bad], num_comparisons=3)

        return IAS_gen, IAS_em, IAS_em_good, IAS_em_bad
```

## Correlation between emotional intensities captured by LexIT and FEEL-IT

```python
import pandas as pd

bing25_lexit = pd.read_csv("../Results/SIGIR_lexicon_queryWise_bingEP.csv")
bing25_feelit = pd.read_csv("../Results/SIGIR_semantic_queryWise_bingEP.csv")

gemma_lexit = pd.read_csv("../Results/SIGIR_lexicon_queryWise_gemmaEP.csv")
gemma_feelit = pd.read_csv("../Results/SIGIR_semantic_queryWise_gemmaEP.csv")

gpt_lexit = pd.read_csv("../Results/SIGIR_lexicon_queryWise_gptEP.csv")
gpt_feelit = pd.read_csv("../Results/SIGIR_semantic_queryWise_gptEP.csv")
```

```python
from scipy import stats

def corr_lex_feel(IAS_name, lexit_vectors, feelit_vectors):
    components = ["positive", "negative", "joy", "sadness", "anger", "fear"]
    results = []
    for component in components:
        stat_val, p_val = stats.pearsonr(lexit_vectors[component], feelit_vectors[component])
        results.append([IAS_name, component, stat_val, p_val, p_val < 0.05])

    return results

bing_corr = corr_lex_feel("Bing'25", bing25_lexit, bing25_feelit)
gemma_corr = corr_lex_feel("Gemma", gemma_lexit, gemma_feelit)
gpt_corr = corr_lex_feel("GPT", gpt_lexit, gpt_feelit)
```

```python
df = pd.DataFrame(bing_corr + gemma_corr + gpt_corr, columns=["IAS", "Vector Component", "Coeff Factor", "p val", "is_significant"])
df.to_csv("../Stats/FEELIT-LEXIT-Correlation.csv", index=False)
df
```
